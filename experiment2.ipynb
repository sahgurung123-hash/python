```python
#10. Write a program to print truth table for bitwise operators (&, |, and ^ operators) 
print("A B A&B A|B A^B")
for A in [0, 1]:
    for B in [0, 1]:
        print(A, B, A & B, A | B, A ^ B)
```

```python
#11. Write a program to find left shift and right shift values of a given number. 
num=int(input("enter a number"))
shift=int(input("enter number of shifts"))
left_shift=num<<shift
right_shift=num>>shift  
print("left shift is",left_shift)
print("right shift is",right_shift)
```

```python
#12.	Using membership operator find whether a given number is in sequence (10,20,56,78,89) 
num=int(input("enter a number"))
sequence = (10, 20, 56, 78, 89)
if num in sequence:
    print("The number is in the sequence.")
else:
    print("The number is not in the sequence.")
```

```python
#13.	Using membership operator find whether a given character is in a string.
char = input("Enter a character: ")
string = input("Enter a string: ")
if char in string:
    print("The character is in the string.")
else:
    print("The character is not in the string.")
```
